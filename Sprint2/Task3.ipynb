```
public class Employee
{
    private string _name;
    private decimal _age;
    private decimal _salary;
    // НОВЫЕ АТРИБУТЫ:
    private string _email;
    private DateTime _hireDate;
    private int _vacationDays;

    public Employee(string name, decimal age, decimal salary, string email, DateTime hireDate)
    {
        _name = name;
        _age = age;
        _salary = salary;
        _email = email;
        _hireDate = hireDate;
        _vacationDays = 28; 
    }

    public string Name
    {
        get { return _name; }
        set 
        {
            if (!string.IsNullOrWhiteSpace(value))
                _name = value;
            else
                throw new ArgumentException("Имя не может быть пустым");
        }
    }

    public decimal Age
    {
        get { return _age; }
        set 
        {
            if (value >= 18 && value <= 70)
                _age = value;
            else
                throw new ArgumentException("Возраст должен быть от 18 до 70 лет");
        }
    }

    public decimal Salary
    {
        get { return _salary; }
        set 
        {
            if (value >= 0)
                _salary = value;
            else
                throw new ArgumentException("Зарплата не может быть отрицательной");
        }
    }

    // НОВЫЕ СВОЙСТВА:
    public string Email
    {
        get { return _email; }
        set
        {
            if (value.Contains("@"))
                _email = value;
            else
                throw new ArgumentException("Email должен содержать символ @");
        }
    }

    public DateTime HireDate
    {
        get { return _hireDate; }
        set { _hireDate = value; }
    }

    public int VacationDays
    {
        get { return _vacationDays; }
        set
        {
            if (value >= 0)
                _vacationDays = value;
            else
                throw new ArgumentException("Количество дней отпуска не может быть отрицательным");
        }
    }

    public virtual void GetInfo()
    {
        Console.WriteLine($"Имя: {Name}. Возраст: {Age}. Зарплата: {Salary}.");
        Console.WriteLine($"Email: {Email}. Дата найма: {HireDate.ToShortDateString()}.");
        Console.WriteLine($"Дней отпуска: {VacationDays}.");
    }

    public virtual void Work()
    {
        Console.WriteLine($"{Name} выполнил свою работу.");
    }

    // ОБНОВЛЕННЫЙ МЕТОД С УЧЕТОМ ДНЕЙ ОТПУСКА:
    public virtual void TakeVacation(int days = 14)
    {
        if (days <= VacationDays)
        {
            VacationDays -= days;
            Console.WriteLine($"{Name} взял отпуск на {days} дней. Осталось дней отпуска: {VacationDays}.");
        }
        else
        {
            Console.WriteLine($"{Name}, у вас недостаточно дней отпуска. Запрошено: {days}, доступно: {VacationDays}.");
        }
    }

    // НОВЫЙ МЕТОД: Рассчитать стаж работы
    public virtual int CalculateTenure()
    {
        return (DateTime.Now - HireDate).Days / 365;
    }

    // НОВЫЙ МЕТОД: Отправить уведомление
    public virtual void SendNotification(string message)
    {
        Console.WriteLine($"Уведомление для {Name} ({Email}): {message}");
    }
}

public class Manager : Employee
{
    private decimal _subordinatesCount;
    private List<Employee> _subordinates;
    // НОВЫЕ АТРИБУТЫ:
    private string _department;
    private decimal _bonus;

    public Manager(string name, decimal age, decimal salary, string email, DateTime hireDate, 
                  decimal subordinatesCount, string department) 
        : base(name, age, salary, email, hireDate)
    {
        _subordinatesCount = subordinatesCount;
        _department = department;
        _subordinates = new List<Employee>();
        _bonus = 0;
    }

    public decimal SubordinatesCount
    {
        get { return _subordinatesCount; }
        set 
        {
            if (value >= 0)
                _subordinatesCount = value;
            else
                throw new ArgumentException("Количество подчиненных не может быть отрицательным");
        }
    }

    // НОВЫЕ СВОЙСТВА:
    public string Department
    {
        get { return _department; }
        set
        {
            if (!string.IsNullOrWhiteSpace(value))
                _department = value;
            else
                throw new ArgumentException("Название отдела не может быть пустым");
        }
    }

    public decimal Bonus
    {
        get { return _bonus; }
        set
        {
            if (value >= 0)
                _bonus = value;
            else
                throw new ArgumentException("Бонус не может быть отрицательным");
        }
    }

    public override void GetInfo()
    {
        base.GetInfo();
        Console.WriteLine($"Отдел: {Department}. Подчиненных в отделе: {SubordinatesCount}.");
        Console.WriteLine($"Управляет {_subordinates.Count} сотрудниками. Бонус: {Bonus}.");
    }

    public override void Work()
    {
        Console.WriteLine($"{Name} управляет отделом '{Department}' " + $"и координирует {_subordinates.Count} сотрудников.");
    }

    // НОВЫЙ МЕТОД: Рассчитать общую зарплату с бонусом
    public decimal CalculateTotalSalary()
    {
        return Salary + Bonus;
    }

    // НОВЫЙ МЕТОД: Провести оценку эффективности
    public void ConductPerformanceReview()
    {
        Console.WriteLine($"{Name} проводит оценку эффективности для {_subordinates.Count} сотрудников:");
        foreach (var subordinate in _subordinates)
        {
            Console.WriteLine($"- Оценивается {subordinate.Name}");
        }
    }

     public void AddSubordinate(Employee employee)
    {
        if (employee != null && !_subordinates.Contains(employee))
        {
            _subordinates.Add(employee);
            Console.WriteLine($"{employee.Name} добавлен в команду менеджера {Name}");
        }
    }

    public void RemoveSubordinate(Employee employee)
    {
        if (_subordinates.Contains(employee))
        {
            _subordinates.Remove(employee);
            Console.WriteLine($"{employee.Name} удален из команды менеджера {Name}");
        }
    }

    public void ListSubordinates()
    {
        Console.WriteLine($"Подчиненные менеджера {Name}:");
        foreach (var subordinate in _subordinates)
        {
            Console.WriteLine($"- {subordinate.Name} ({subordinate.GetType().Name})");
        }
    }

    public void ConductMeeting()
    {
        Console.WriteLine($"{Name} проводит собрание с командой:");
        foreach (var subordinate in _subordinates)
        {
            Console.WriteLine($"- {subordinate.Name} присутствует на собрании");
        }
    }
}

public class Developer : Employee
{
    private string _specialization;
    // НОВЫЕ АТРИБУТЫ:
    private string _techStack;
    private bool _hasCertification;
    private int _completedProjects;

    public Developer(string name, decimal age, decimal salary, string email, DateTime hireDate, string specialization, string techStack) 
        : base(name, age, salary, email, hireDate)
    {
        _specialization = specialization;
        _techStack = techStack;
        _hasCertification = false;
        _completedProjects = 0;
    }

    public string Specialization
    {
        get { return _specialization; }
        set 
        {
            if (!string.IsNullOrWhiteSpace(value))
                _specialization = value;
            else
                throw new ArgumentException("Специализация не может быть пустой");
        }
    }

    // НОВЫЕ СВОЙСТВА:
    public string TechStack
    {
        get { return _techStack; }
        set
        {
            if (!string.IsNullOrWhiteSpace(value))
                _techStack = value;
            else
                throw new ArgumentException("Стек технологий не может быть пустым");
        }
    }

    public bool HasCertification
    {
        get { return _hasCertification; }
        set { _hasCertification = value; }
    }

    public int CompletedProjects
    {
        get { return _completedProjects; }
        set
        {
            if (value >= 0)
                _completedProjects = value;
            else
                throw new ArgumentException("Количество завершенных проектов не может быть отрицательным");
        }
    }

    public override void GetInfo()
    {
        base.GetInfo();
        Console.WriteLine($"Специализация: {Specialization}. Стек технологий: {TechStack}.");
        Console.WriteLine($"Сертификация: {(HasCertification ? "Да" : "Нет")}. Завершенные проекты: {CompletedProjects}.");
    }

    public override void Work()
    {
        Console.WriteLine($"{Name} ({Specialization}) пишет код используя {TechStack}.");
    }

    // НОВЫЙ МЕТОД: Получить сертификацию
    public void GetCertified(string certificationName)
    {
        HasCertification = true;
        Console.WriteLine($"{Name} получил сертификацию: {certificationName}");
    }

    // НОВЫЙ МЕТОД: Завершить проект
    public void CompleteProject(string projectName, bool isSuccess = true)
    {
        if (isSuccess)
        {
            CompletedProjects++;
            Console.WriteLine($"{Name} успешно завершил проект: '{projectName}'. Всего завершенных проектов: {CompletedProjects}");
        }
        else
        {
            Console.WriteLine($"{Name} столкнулся с проблемами в проекте: '{projectName}'");
        }
    }
}

// СЛОЖНОЕ (МНОГОУРОВНЕВОЕ) НАСЛЕДОВАНИЕ: TeamLead -> Developer -> Employee
public class TeamLead : Developer
{
    private string _teamName;
    private int _teamSize;
    private List<Developer> _teamMembers;

    public TeamLead(string name, decimal age, decimal salary, string email, DateTime hireDate, string specialization, string techStack, string teamName, int teamSize)
        : base(name, age, salary, email, hireDate, specialization, techStack)
    {
        _teamName = teamName;
        _teamSize = teamSize;
        _teamMembers = new List<Developer>();
    }

    public string TeamName
    {
        get { return _teamName; }
        set
        {
            if (!string.IsNullOrWhiteSpace(value))
                _teamName = value;
            else
                throw new ArgumentException("Название команды не может быть пустым");
        }
    }

    public int TeamSize
    {
        get { return _teamSize; }
        set
        {
            if (value >= 0)
                _teamSize = value;
            else
                throw new ArgumentException("Размер команды не может быть отрицательным");
        }
    }

    public override void GetInfo()
    {
        base.GetInfo();
        Console.WriteLine($"Команда: {TeamName}. Размер команды: {TeamSize}.");
        Console.WriteLine($"Руководит {_teamMembers.Count} разработчиками.");
    }

    public override void Work()
    {
        Console.WriteLine($"{Name} руководит командой '{TeamName}' и участвует в разработке.");
    }

    // МЕТОДЫ УПРАВЛЕНИЯ КОМАНДОЙ:
    public void AssignTask(Developer developer, string task)
    {
        if (_teamMembers.Contains(developer))
        {
            Console.WriteLine($"{Name} назначил задачу '{task}' разработчику {developer.Name}");
        }
        else
        {
            Console.WriteLine($"{developer.Name} не является членом команды {TeamName}");
        }
    }

    public void ConductCodeReview()
    {
        Console.WriteLine($"{Name} проводит код-ревью для команды {TeamName}:");
        foreach (var member in _teamMembers)
        {
            Console.WriteLine($"- Проверка кода {member.Name}");
        }
    }

    public void AddTeamMember(Developer developer)
    {
        if (!_teamMembers.Contains(developer))
        {
            _teamMembers.Add(developer);
            Console.WriteLine($"{developer.Name} добавлен в команду {TeamName}");
        }
    }

    public void RemoveTeamMember(Developer developer)
    {
        if (_teamMembers.Contains(developer))
        {
            _teamMembers.Remove(developer);
            Console.WriteLine($"{developer.Name} удален из команды {TeamName}");
        }
    }
}

// ИНТЕРФЕЙСЫ ДЛЯ МНОЖЕСТВЕННОГО НАСЛЕДОВАНИЯ
public interface ITrainer
{
    void ConductTraining(string topic);
    void CreateTrainingMaterials();
    int TrainingSessionsConducted { get; set; }
}

public interface IProjectManager
{
    void CreateProjectPlan(string projectName);
    void TrackProjectProgress();
    int ManagedProjects { get; set; }
}

// КЛАСС С МНОЖЕСТВЕННЫМ НАСЛЕДОВАНИЕМ (через интерфейсы)
public class TechnicalLead : TeamLead, ITrainer, IProjectManager
{
    private int _trainingSessionsConducted;
    private int _managedProjects;

    public TechnicalLead(string name, decimal age, decimal salary, string email, DateTime hireDate, string specialization, string techStack, string teamName, int teamSize)
        : base(name, age, salary, email, hireDate, specialization, techStack, teamName, teamSize)
    {
        _trainingSessionsConducted = 0;
        _managedProjects = 0;
    }

    // Реализация ITrainer
    public int TrainingSessionsConducted
    {
        get { return _trainingSessionsConducted; }
        set { _trainingSessionsConducted = value; }
    }

    public void ConductTraining(string topic)
    {
        TrainingSessionsConducted++;
        Console.WriteLine($"{Name} проводит тренинг по теме: '{topic}'. Всего проведено тренингов: {TrainingSessionsConducted}");
    }

    public void CreateTrainingMaterials()
    {
        Console.WriteLine($"{Name} создает учебные материалы для команды {TeamName}");
    }

    // Реализация IProjectManager
    public int ManagedProjects
    {
        get { return _managedProjects; }
        set { _managedProjects = value; }
    }

    public void CreateProjectPlan(string projectName)
    {
        ManagedProjects++;
        Console.WriteLine($"{Name} создает план проекта '{projectName}'. Всего управляемых проектов: {ManagedProjects}");
    }

    public void TrackProjectProgress()
    {
        Console.WriteLine($"{Name} отслеживает прогресс по всем проектам команды {TeamName}");
    }

    public override void GetInfo()
    {
        base.GetInfo();
        Console.WriteLine($"Проведено тренингов: {TrainingSessionsConducted}. Управляемых проектов: {ManagedProjects}.");
    }
}



Console.WriteLine("=== ПРОСТОЕ НАСЛЕДОВАНИЕ ===");
        
Manager manager = new Manager("Ник", 35, 80000, "nik@company.com", new DateTime(2018, 5, 15), 15, "IT");
        
Developer developer1 = new Developer("Анна", 28, 60000, "anna@company.com", new DateTime(2020, 3, 10), "Backend", "C#, .NET, SQL");
        
Developer developer2 = new Developer("Максим", 32, 65000, "maxim@company.com", new DateTime(2019, 8, 22), "Frontend", "JavaScript, React");

manager.AddSubordinate(developer1);
manager.AddSubordinate(developer2);
        
developer1.GetCertified("Microsoft Certified Developer");
developer1.CompleteProject("Система аналитики");
developer1.TakeVacation(10);
        
manager.Bonus = 5000;
Console.WriteLine($"Общая зарплата менеджера с бонусом: {manager.CalculateTotalSalary()}");

Console.WriteLine("\n=== СЛОЖНОЕ НАСЛЕДОВАНИЕ ===");
        
TeamLead teamLead = new TeamLead("Алексей", 38, 90000, "alexey@company.com", new DateTime(2017, 2, 14), "FullStack", "C#, Angular, Azure", "Alpha Team", 5);
        
teamLead.AddTeamMember(developer1);
teamLead.AssignTask(developer1, "Оптимизация базы данных");
teamLead.ConductCodeReview();
teamLead.CompleteProject("Мобильное приложение", true);

Console.WriteLine("\n=== МНОЖЕСТВЕННОЕ НАСЛЕДОВАНИЕ ===");
        
TechnicalLead techLead = new TechnicalLead("Мария", 40, 95000, "maria@company.com", new DateTime(2015, 11, 5), "Architect", ".NET, Microservices, AWS", "Innovation Team", 8);
        
techLead.ConductTraining("Введение в микросервисы");
techLead.CreateProjectPlan("Облачная платформа");
techLead.TrackProjectProgress();
techLead.CreateTrainingMaterials();
        
techLead.GetInfo();

Console.WriteLine($"\nСтаж работы Марии: {techLead.CalculateTenure()} лет");
techLead.SendNotification("Не забудьте подготовить отчет за квартал");
```
